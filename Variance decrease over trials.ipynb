```python
#################################################################################################
## Import Section & Session creation

from data_session import Session
import numpy as np
import matplotlib.pyplot as plt

file_name = 'V1_SESSION_B20141210_1.mat'

s = Session(file_name,normalized=False)
s1 = Session(file_name,normalized=True)
```

```python
#Variance decrease over trials

for i in range(5):
    a = s.get_data("seen")[i]
    c = s.smoother(s.var_over(a,channels=True,time=True),10)
    plt.plot(c)
    plt.title("Variance over trials for contrast #"+str(i+1) + "(Data not normalized)")
    plt.figure()
    a1 = s1.get_data("seen")[i]
    c1 = s1.smoother(s1.var_over(a1,channels=True,time=True))
    plt.plot(c1)
    plt.title("Variance over trials for contrast #"+str(i+1) + "(Data normalized)")
    plt.figure()
plt.show()
```
